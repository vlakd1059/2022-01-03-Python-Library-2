### 선그래프

```python
# plot 함수 - 일직선으로 되어있는 Lineplot그리기
y=[2,4,6]

plt.plot(y)  # 그래프를 그리기
plt.show()  # 그래프를 출력
```

```python
# x축 설정하여 그래프 그리기
x=[1,2,3]
y=[2,4,6]

plt.plot(x, y)
plt.show()
```

```python
# line style -> ls
# 그려지는 라인의 스타일 지정
plt.plot(x,y, ls='--')
plt.show()
```

```python
x=[2,4,6,8]
y=[10,13,16,20]
y2=[10,20,16,13]
plt.plot(x,y)
plt.plot(x,y2, ls =':')
plt.show()
```

```python
plt.plot(x,y)
plt.show()
plt.plot(x,y2)
plt.show()
```

```python
# marker
plt.plot(x,y, marker='o')
plt.show()
```

```python
# line width ->lw
# 라인 두께 지정
plt.plot(x,y, lw='5')
plt.show()
```

```python
# 선 색지정
# 마커 색지정
# ->  markerfacecolor -> mfc
plt.plot(x,y, color='r', marker='o', mfc='b')
plt.show()
```

```python
x=[0,1,2,3,4,5,6]
y=[1,4,5,8,9,5,3]

plt.plot(x,y, color='g', marker='o',ls=':')
plt.show()
```

```python
x=[0,1,2,3,4,5,6]
y=[1,4,5,8,9,5,3]

plt.plot(x,y, color='r', marker='d',ls=':', mfc='b')
plt.show()
```

```python
import numpy as np
x= np.arange(7)
y=[1,4,5,8,9,5,3]
plt.plot(x,y, lw=5, marker='o',ls='--', ms=15,  mfc='r', mec='g', mew=5)
plt.show()
```
